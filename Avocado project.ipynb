# To check Correlation

```python
df_cor=df.corr()
df_cor
```

Observation:

Here we can see the correlation between the columns

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_cor,annot=True)
```

Observation:

1.Here we can see that no column has very high or very low correlation with target variable 'Average Price', so all are equally important.So it's better to keep all the columns.

2.Total bag, Small bag ,Large bag,4770,4225 are highly correlated.

3.The graph shows high multi-colinearity between independent variables.Few are highlighted below:
    Total Volume & 4046, Total Volume & 4225, Total Volume & Large Bags, Small Bags & Total Bags.

# Data Visualization

Univariate Analysis

```python
df.columns
```

```python
plt.figure(figsize=(15,15))
df['AveragePrice'].plot.bar()
plt.ylabel('Average Price')
plt.xlabel('No. of Rows')
plt.xticks(rotation=45)
plt.title('Average Price decripition')
plt.show()
```

```python
plt.figure(figsize=(15,15))
df['Total Volume'].plot.bar()
plt.ylabel('Total Volume')
plt.xlabel('No. of Rows')
plt.xticks(rotation=45)
plt.title('Total Volume decripition')
plt.show()
```

```python
df['Total Volume'].plot.hist(bins=20)
plt.title('Histogram of Total Volume')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.violinplot(df['Total Volume'])
plt.show()
```

Observation:

There is a skewness in the Total Volume.
